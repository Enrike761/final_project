```python
import requests
import pandas as pd
import numpy as np

API_KEY = "AIzaSyCK1HWjYgZGTVm-AVUIQZfiaMErUCcfj-A" # my API key

# ubicacion = "37.389257, -5.921089" # let´s add several locations because we are not getting the desired amount of results

ubicaciones = [
    (37.389257, -5.921089),  
    (37.391000, -5.923500),  
    (37.387000, -5.918500),  
    (37.392500, -5.920000),  
    (37.385500, -5.922500)   
]


radio = 1800  # the radio in meters need to be small for better accuracy

tipo = "supermarket|grocery_or_supermarket|food|store" # here we include all tipes of establishments, the problem is that we are only getting info of big establishments, so I will add a text search for more accuracy

palabras_clave = "supermarket OR chucheria OR tienda de comestibles OR minimarket OR ultramarinos OR gasolinera OR autoservicio OR chino OR tienda de alimentacion"

url = url = f"https://maps.googleapis.com/maps/api/place/textsearch/json?query={palabras_clave}&location={ubicaciones}&radius={radio}&key={API_KEY}"
#f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={ubicacion}&radius={radio}&type={tipo}&key={API_KEY}"

response = requests.get(url)
data = response.json()

lugares = []
for lugar in data.get("results", []):
    nombre = lugar.get("name")
    direccion = lugar.get("vicinity")
    lat = lugar["geometry"]["location"]["lat"]
    lon = lugar["geometry"]["location"]["lng"]
    influencia = 1000

    lugares.append([nombre, direccion, lat, lon, influencia])

df = pd.DataFrame(lugares, columns=["Tienda", "Dirección", "Latitud", "Longitud", "Influencia"])

df.to_csv("supermercados3.csv", index=False)

print("Archivo supermercados.csv generado con éxito.")
```
